```python
# Needing to inport the Matplotlit and Pyplot libs. 
import matplotlib.pyplot as plt

# Allows us to see the plots in the notebook 

%matplotlib inline 
x = list(range(0,10))
y = list (range(-10,0))

plt.plot(x, y)
```

```python
#Creating some data sets with python arrays.
a = [0, -100, 25, 67, -323] 
b = [0, 3, 7, 3, 9]
plt.plot(a, b)
```

```python
#Focusing in on the triangle area of the graph 
plt.axis([-50, 80, 2, 8])
plt.plot(a, b)
```

```python
# Naming the graph and axis. 
# Focusing on the triangular section of the graph
# Naming our x axis values, changing them from the numerical values to chars
plt.title("Triangle Graph")
plt.xlabel("Array A")
plt.ylabel("Array B")
plt.axis([-50, 80, 2, 8])
plt.xticks((-40, -20, 0, 20, 40, 60, 80), ('h', 'e', 'l', 'l', 'o', 'o', 'o'))
plt.plot(a, b, color="red")
```

```python
# Creating a histogram from our data to see a different type of how we can distribute the data
# Notice all of the x values are in blue, positive values
# All of the y values are in orange, negative values
plt.hist(x)
plt.hist(y)
```
